```python

from scipy import optimize
import matplotlib.pyplot as plt

x=np.array(dataset["Tiempo (H)"])

f1=lambda x,a,b,c: a*pow(x,3)+b*pow(x,2)+c*x+2

y=np.array(dataset["Velocidad (m/s)"])

coef=optimize.curve_fit(f1,x,y)[0]

plt.figure(figsize=(16,4.5))

plt.subplot(1,2,1)
plt.scatter(x,y,color="#ABE338")
plt.title("Gráfica Velocidad/Tiempo")
plt.xlabel("Tiempo")
plt.ylabel("Velocidad")
plt.grid()

plt.subplot(1,2,2)
plt.scatter(x,y, color="#ABE338")
plt.title("Regresion Cubica")
plt.xlabel("Tiempo")
plt.ylabel("Velocidad")
plt.plot(x,f1(x,coef[0],coef[1],coef[2]))
plt.grid()
plt.show()
```
