```python
import sklearn as sk
import numpy as np
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
x, y = make_blobs ( n_samples = 500 , centers = 4, n_features = 5 , random_state = ( 8 + 45))
plt.scatter( x.T[0], x.T[1] , c = y )
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

X, xt, Y, yt = train_test_split(x, y, random_state = 24)

model = MLPClassifier (
    hidden_layer_sizes = [30, 35, 35, 30],
    activation = "logistic",
    solver = "adam",
    random_state = 99,
    verbose = True
)

model.fit(X,Y)
print(model.score(X,Y))
print(model.score(xt,yt))
```

```python
p = model.predict(xt)
plt.subplot(1, 2, 1)
plt.scatter(X.T[0], X.T[1], c = Y)
plt.subplot(1, 2, 2)
plt.scatter(xt.T[0], xt.T[1], c = yt)
```
